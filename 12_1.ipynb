```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.fashion_mnist.load_data()

N = 30000

x_train = x_train[:N] / 255
x_test = x_test[:N] / 255

y_train = y_train[:N]
y_test = y_test[:N]


x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
def plot_history(hst):
    plt.figure(figsize=(13, 4))

    plt.subplot(1, 2, 1)
    plt.plot(hst.history['loss'], label='train')
    plt.plot(hst.history['val_loss'], label='test')
    plt.title('Loss')

    plt.subplot(1, 2, 2)
    plt.plot([round(100*e, 2) for e in hst.history['sparse_categorical_accuracy']], label='train')
    plt.plot([round(100*e, 2) for e in hst.history['val_sparse_categorical_accuracy']], label='test')
    plt.title('Accuracy')

    plt.ylim(0, 100)
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
collect_opt = {}
collect_act = {}

optimizers_list = [
    tf.keras.optimizers.Adam,
    tf.keras.optimizers.Adagrad,
    tf.keras.optimizers.Adamax,
]

activation_list = [
    'relu',
    'softplus',
    'sigmoid'
]

for opt in optimizers_list:
  for act in activation_list:
      model = tf.keras.Sequential([
          tf.keras.layers.Flatten(input_shape=(28, 28)),
          tf.keras.layers.Dense(128, activation=act),
          tf.keras.layers.Dense(64, activation=act),
          tf.keras.layers.Dense(10)
      ])

      model.compile(
          loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
          optimizer=opt(learning_rate=0.001),
          metrics=[tf.keras.metrics.SparseCategoricalAccuracy()]
      )

      history = model.fit(x_train, y_train, epochs=5, validation_data=(x_test, y_test))
      print(f"{opt.__name__} with activation {act}")
      plot_history(history)

      if opt.__name__ not in collect_opt:
        collect_opt[opt.__name__] = {}
      if act not in collect_act:
        collect_act[act] = {}

      collect_opt[opt.__name__][act] = history.history['val_sparse_categorical_accuracy']
      collect_act[act][opt.__name__] = history.history['val_sparse_categorical_accuracy']
```

```python
for collection in collect_opt:
    print(f"{collection}")
    display(pd.DataFrame(collect_opt[collection]).T.style.background_gradient(cmap='viridis'))
    print(f"\n")
```

```python
collect_ep = {}

epoch_list = [5, 10, 15, 20]

for epoch in epoch_list:
  model = tf.keras.Sequential([
      tf.keras.layers.Flatten(input_shape=(28, 28)),
      tf.keras.layers.Dense(128, activation='relu'),
      tf.keras.layers.Dense(64, activation='relu'),
      tf.keras.layers.Dense(10)
  ])

  model.compile(
      loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
      optimizer=tf.keras.optimizers.Adam(0.001),
      metrics=[tf.keras.metrics.SparseCategoricalAccuracy()]
  )

  history = model.fit(x_train, y_train, epochs=epoch, validation_data=(x_test, y_test))
  plot_history(history)
  collect_ep[epoch] = history.history['val_sparse_categorical_accuracy']
```

```python
from IPython.display import display

for epoch, accuracy in collect_ep.items():
    print(f"Results for {epoch} epochs:")

    df = pd.DataFrame({epoch: accuracy}).T
    display(df.style.background_gradient(cmap='viridis'))
    print("\n")
```

```python
collect_batch = {}

batch_list = [16, 32, 64]

for batch in batch_list:
    model = tf.keras.Sequential([
        tf.keras.layers.Flatten(input_shape=(28, 28)),
        tf.keras.layers.Dense(128, activation='relu'),
        tf.keras.layers.Dense(64, activation='relu'),
        tf.keras.layers.Dense(10)
    ])

    model.compile(
        loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
        optimizer=tf.keras.optimizers.Adam(0.001),
        metrics=[tf.keras.metrics.SparseCategoricalAccuracy()]
    )

    history = model.fit(x_train, y_train, epochs=20, batch_size=batch, validation_data=(x_test, y_test))
    plot_history(history)

    collect_batch[batch] = history.history['val_sparse_categorical_accuracy']
```

```python
for batch, accuracy in collect_batch.items():
    print(f"Results for batch size {batch}:")
    df = pd.DataFrame({batch: accuracy}).T
    display(df.style.background_gradient(cmap='viridis'))
    print("\n")
```
